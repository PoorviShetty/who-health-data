```python
# assuming all data is downloaded from online already
import pandas as pd
import os
import re
```

```python
ghe_death = pd.read_excel('original_data/rep_ghe_death_sex/data.xlsx', sheet_name='Sheet1', header=0)
ghe_death.head()
```

```python
nut = pd.read_excel('original_data/rep_nut/data.xlsx', sheet_name='Sheet1', header=0)
nut.head()
```

```python
rmnch = pd.read_excel('original_data/rep_rmnch/data.xlsx', sheet_name='Sheet1', header=0)
rmnch.head()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder \
    .appName("WHO Data Analysis") \
    .getOrCreate()
```

```python
rmnch_spark_df = spark.createDataFrame(rmnch)

year_options = [2000, 2005, 2010, 2015, 2019]
dimensions_options = ["Child's age (2 groups) (0-5)", 'Economic status (wealth quintile)', 'Education (3 groups)', 'Place of residence', 'Sex']

rmnch_df = rmnch_spark_df.filter(col('dimension').isin(dimensions_options))
rmnch_df = rmnch_df.filter(col('subgroup') != 'Male')
rmnch_df = rmnch_df.filter(col('date').isin(year_options))

rmnch_df.show()
```

```python
nut_spark_df = spark.createDataFrame(nut)

year_options = [2000, 2005, 2010, 2015, 2019]
dimensions_options = ["Child's age (2 groups) (0-5)", 'Economic status (wealth quintile)', 'Education (3 groups)', 'Place of residence', 'Sex']

nut_df = nut_spark_df.filter(col('dimension').isin(dimensions_options))
nut_df = nut_df.filter(col('subgroup') != 'Male')
nut_df = nut_df.filter(col('date').isin(year_options))

nut_df.show()
```

```python
year_options = [2000, 2005, 2010, 2015, 2019]
suffix_options = ['_0000', '_0010', '_0600', '_1510']

# ghe_daly_spark_df = spark.createDataFrame(ghe_daly)
ghe_death_spark_df = spark.createDataFrame(ghe_death)
# ghe_yld_spark_df = spark.createDataFrame(ghe_yld)
# ghe_yll_spark_df = spark.createDataFrame(ghe_yll)



def filter_and_select(df, suffix_options):
    filtered_df = df.filter(col('subgroup') != 'Male').filter(col('date').isin(year_options))
    condition = (
        col('indicator_abbr').endswith('_0000') |
        col('indicator_abbr').endswith('_0010') |
        col('indicator_abbr').endswith('_0600') |
        col('indicator_abbr').endswith('_1510')
    )
    # Filter based on indicator_abbr suffixes
    # condition = col('indicator_abbr').endswith(tuple(suffix_options))
    filtered_df = filtered_df.filter(condition)

    return filtered_df

# ghe_daly_df = filter_and_select(ghe_daly_spark_df, suffix_options)
ghe_death_df = filter_and_select(ghe_death_spark_df, suffix_options)
# ghe_yld_df = filter_and_select(ghe_yld_spark_df, suffix_options)
# ghe_yll_df = filter_and_select(ghe_yll_spark_df, suffix_options)

ghe_death_df.show()
```

```python
rmnch_df.toPandas().to_csv('filtered_data/rmnch.csv')
nut_df.toPandas().to_csv('filtered_data/nut.csv')
# oecd_hc_df.toPandas.to_csv('filtered_data/oecd_hc.csv')
# ghe_daly_df.toPandas.to_csv('filtered_data/ghe_daly.csv')
ghe_death_df.toPandas().to_csv('filtered_data/ghe_death.csv')
# ghe_yld_df.toPandas.to_csv('filtered_data/ghe_yld.csv')
# ghe_yll_df.toPandas.to_csv('filtered_data/ghe_yll.csv')
```

```python
# SPLITTING NUT

df = nut_spark_df

df.printSchema()
df.show(5)

indicator_abbr_values = df.select("indicator_abbr").distinct().collect()

df_dict = {}

for row in indicator_abbr_values:
    indicator_abbr_value = row['indicator_abbr']
    df_filtered = df.filter(df["indicator_abbr"] == indicator_abbr_value)
    df_dict[indicator_abbr_value] = df_filtered


for key, df_partition in df_dict.items():
    if not os.path.exists('filtered_data/nut'):
        os.makedirs('filtered_data/nut')
    df_partition.toPandas().to_csv(f"filtered_data/nut/nut_partitioned_{key}.csv", header=True)
    dimension_values = df_partition.select("dimension").distinct().collect()
    df_dict_dim = {}
    for row in dimension_values:
        dimension_value = row['dimension']
        df_filtered_dim = df_partition.filter(df_partition["dimension"] == dimension_value)
        df_dict_dim[dimension_value] = df_filtered_dim


        for dim_key, df_partition_dim in df_dict_dim.items():
            df_partition_dim.toPandas().to_csv(f"filtered_data/nut/nut_{key}_partitioned_{dim_key}.csv", header=True)
```

```python
# SPLITTING RMNCH: Attempt 1

df = rmnch_spark_df

df.printSchema()
df.show(5)

indicator_values = df.select("indicator_name").distinct().collect()

df_dict = {}

for row in indicator_values:
    indicator_value = row['indicator_name']
    df_filtered = df.filter(df["indicator_name"] == indicator_value)
    df_dict[indicator_value] = df_filtered


for key, df_partition in df_dict.items():
    if not os.path.exists('filtered_data/rmnch'):
        os.makedirs('filtered_data/rmnch')
    key = re.sub(r'[^\w\s]', '', key).replace(' ', '_')
    df_partition.toPandas().to_csv(f"filtered_data/rmnch/rmnch_partitioned_{key}.csv", header=True)
```

```python
# SPLITTING RMNCH: Attempt 2

women_indicators = ['Women (aged 20-24 years) who gave birth before age 18 (%)',
    'Women with a birth in the past five years suffering from night blindness during pregnancy (adjusted) (%)',
    'Women with a birth in the past five years who received a vitamin A dose in the first two months after delivery (%)',
    'Women with a birth in the past five years who took deworming medication during pregnancy of last birth (%)',
    'Women with a birth in the past five years who took iron tablets or syrup for 90+ days (%)',
    'Mothers who received postnatal care within two days of giving birth (%)',
    'Deliveries in a health facility (%)']

newborn_indicators = ['Newborns who received postnatal care within two days of delivery (%)',
    'Children given deworming medication in past 6 months (%)',
    'Children given iron supplements in past 7 days (%)',
    'Severe anaemia prevalence among children (%)',
    'Under-five mortality rate (deaths per 1000 live births)',
    'Full immunization coverage among one-year-olds (%)',
    'Predominantly breastfeeding (0 to 5 months) (%)']


# SPLITTING RMNCH
df = spark.read.csv("filtered_data/rmnch.csv", header=True, inferSchema=True)

filtered_df = df.filter(col('indicator_name').isin(women_indicators))
filtered_df.toPandas().to_csv(f"filtered_data/rmnch/rmnch_women_partitioned.csv", header=True)

filtered_df = df.filter(col('indicator_name').isin(newborn_indicators))
filtered_df.toPandas().to_csv(f"filtered_data/rmnch/rmnch_newborn_partitioned.csv", header=True)
```
